### (M) 78. Subsets

86 / 90

```python
nums = [1, 2, 3]

class Solution(object):
    def subsets(self, nums):
        self.ans = []

        def chooseOrNot(nums, subset):
          if len(nums) == 1:
            self.ans.extend([subset + nums, subset])
            return

          chooseOrNot(nums[1:], subset + [nums[0]])
          chooseOrNot(nums[1:], subset)

        chooseOrNot(nums, [])
        return self.ans

Solution().subsets(nums)
```

### (M) 148. Sort List â˜…

Follow up: Can you sort the linked list in O(n logn) time and O(1) memory (i.e. constant space)?

image.png

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

head = ListNode(4)
head.next = ListNode(2)
head.next.next = ListNode(1)
head.next.next.next = ListNode(3)
```

Time complexity : O(nlongn)<br>
Space complexity : O(n)<br><br>80 / 17

```python
class Solution(object):
    def sortList(self, head):
        if head == None:
          return None

        cursor = head
        newhead = ListNode(-1)

        values = []
        while cursor != None:
          values.append(cursor.val)
          cursor = cursor.next

        values = sorted(values)
        cursor = newhead
        for val in values:
          cursor.next = ListNode(val)
          cursor = cursor.next

        return newhead.next

Solution().sortList(head)
```

### (E) 404. Sum of Left Leaves

image.png

```python
class TreeNode(object):
     def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
```

1. DFS (preorder traversal)



34 / 100

```python
class Solution(object):
    def sumOfLeftLeaves(self, root):
        
        stack = [root]
        leftSum = 0

        while stack:
          cursor = stack.pop()

          if cursor.right != None: # if cursor has a right child
            stack.append(cursor.right)
          if cursor.left != None: # if cursor has a left child
            stack.append(cursor.left)
            if (cursor.left.right == None) and (cursor.left.left == None): # If the left child has no children -> left leaf
              leftSum+=cursor.left.val

        return leftSum
        
```

### (M) 198. House Robber â˜…

Independent set problem using DP - 88 / 43

```python
nums = [2,7,9,3,1]

class Solution(object):
    def rob(self, nums):
        if len(nums) <= 2: return max(nums)

        dp = [nums[0], max(nums[:2])] + [0] * (len(nums) - 2)
        for i in range(2, len(nums)):
          dp[i] = max(dp[i-1], dp[i-2] + nums[i])

        return dp[-1]

Solution().rob(nums)
```

### (E) 203. Remove Linked List Elements

image.png

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(6)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(4)
head.next.next.next.next.next = ListNode(5)
head.next.next.next.next.next.next = ListNode(6)
```

```python
val = 6
```

Straight-forward idea : 1. Count the values except for targets 2. Connect them - 47 / 10

```python
class Solution(object):
    def removeElements(self, head, val):
        if head == None: return None

        mystack = []
        cursor = head
        while cursor != None:
          if cursor.val != val:
            mystack.append(cursor.val)
          cursor = cursor.next

        if mystack == []: return None

        newhead = ListNode(mystack[0])
        newCursor = newhead
        for i in mystack[1:]:
          newCursor.next = ListNode(i)
          newCursor = newCursor.next
        
        return newhead
        
Solution().removeElements(head,val)
```

### (M) 22. Generate Parentheses

New attempt: instead of creating a real stack, define a pseudostack called 'Num_stack' whose value is integer for memory efficiency - 80 / 70

```python
n = 1
class Solution(object):
    def generateParenthesis(self, n):
        res = []
        Num_stack = 1
        self.n = n

        def pairCount(open_cnt, close_cnt, combi, numStack):
          if open_cnt < 0 or close_cnt < 0 :
            return

          if numStack == 0: 
            
            if open_cnt == 0 == close_cnt:
              res.append(combi)
              return
            
          if numStack < 0:  
            if (open_cnt != self.n) or (close_cnt != self.n):
              return
            
          pairCount(open_cnt - 1, close_cnt, combi + '(', numStack + 1)
          pairCount(open_cnt, close_cnt - 1, combi + ')', numStack - 1)

        pairCount(self.n - 1, self.n, '(', Num_stack)
        return res 

Solution().generateParenthesis(n)
        
```

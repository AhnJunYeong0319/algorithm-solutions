### (M) 62. Unique Paths / ★

23 / 44

```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        grid = [ [1 for j in range(n)] for i in range(m)]

        if (m <= 0) or (n <= 0):
            return 0
        if (m == 1) or (n == 1):
            return 1


        for i in range(1, m):
            for j in range(1, n):

                grid[i][j] = grid[i-1][j] + grid[i][j-1]

        return grid[m - 1][n - 1]
```

```python
class Solution(object):
    def uniquePaths(self, m, n):
        dp = [1] * n
        for i in range(1, m):
            for j in range(1, n):
                dp[j] = dp[j - 1] + dp[j]
        return dp[-1] if m and n else 0
```

### (E) 205. Isomorphic Strings

96 / 42

```python
class Solution(object):
    def isIsomorphic(self, s, t):
        hash = {}
        for idx in range(len(s)):
          if s[idx] in hash: # if s[idx] is redundant
            if hash[s[idx]] == t[idx]:
              continue
            else:
              return False
          
          # case that s[idx] is unique while t[idx] is redundant
          if t[idx] in hash.values():
            return False
          else:  
            hash[s[idx]] = t[idx]
        
        return True
```

### (M) 394. Decode String / ★★

62 / 54

```python
class Solution(object):
    def decodeString(self, s):
        Stack = []
        curNum = 0
        String = ''
        
        for char in s:

            if char == '[':
                Stack.append(String)
                Stack.append(curNum)
                String = ''
                curNum = 0

            elif char == ']':
                num = Stack.pop()
                prevString = Stack.pop()
                String = prevString + num*String

            elif char.isdigit():
                curNum = curNum*10 + int(char)

            else:
                String += char
        return String
```
